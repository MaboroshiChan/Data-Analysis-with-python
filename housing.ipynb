```python
from sklearn.cross_validation import train_test_split
import numpy as np
import pandas as pd
```

```python
sample_submission = pd.read_csv("all/sample_submission.csv")
sample_submission.head()
```

```python
train = pd.read_csv("all/train.csv")
train.head()
```

```python
test = pd.read_csv("all/test.csv")
test.head()
```

```python
train = train[['MSSubClass','LotFrontage','LotArea','OverallQual','YrSold','SalePrice']]
test = test[['MSSubClass','LotFrontage','LotArea','OverallQual','YrSold']]
train.head()
```

```python
median_train = train['LotFrontage'].median()
median_test = test['LotFrontage'].median()
train['LotFrontage'].fillna(median_train, inplace=True)
test['LotFrontage'].fillna(median_test, inplace=True)
```

```python
y_train =pd.Series(train["SalePrice"])
train.drop(["SalePrice"], axis=1, inplace=True)
train.head()
```

```python
X_train = train.values
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
test.head()
```

```python
X_test = test.values
model.predict(X_test)
```

```python
result = np.array(model.predict(X_test))
result = result.astype(int)
```

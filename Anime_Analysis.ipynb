```python
import numpy as np
ranges = [i for i in np.arange(0, 11,0.5)]
C = anime.groupby(pd.cut(anime.rating, ranges)).count()
```

```python
C
```

```python

plt.bar([i for i in np.arange(0,11,0.5)[:-1]],C.anime_id.values, color='g', label='anime')
plt.xlabel('scores')
plt.ylabel('frequency')
plt.title('Rating frequency')
plt.legend()
plt.show()
```

```python
episodes = []
for each in anime.episodes.values:
    try:
        episodes.append(int(each))
    except:
        continue
```

```python
episodes = np.array(episodes)
episodes
```

```python
episodes = pd.DataFrame({'episodes':episodes})
episodes.head()
```

```python
ep_ranges = [i for i in np.arange(1,500,5)]
epi_grouped = episodes.groupby(pd.cut(episodes.episodes, ep_ranges)).count()
epi_grouped.head()
```

```python
plt.scatter(np.arange(0,495,5),epi_grouped.episodes.values, color='g', label='episodes')
plt.xlabel('episodes')
plt.ylabel('frequency')
plt.title('episodes frequency')
plt.legend()
plt.show()
```
